```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import yfinance as yf

stock_symbol = "RELIANCE.NS"

stock_data = yf.download(stock_symbol, start="2023-02-01", end="2023-05-19")

closing_prices = stock_data["Close"].values

window = 20
sma = np.zeros_like(closing_prices)
for i in range(window, len(closing_prices)):
    sma[i] = np.mean(closing_prices[i-window:i])

alpha = 2 / (window + 1)  
ema = np.zeros_like(closing_prices)
ema[window] = np.mean(closing_prices[:window])
for i in range(window + 1, len(closing_prices)):
    ema[i] = (closing_prices[i] - ema[i-1]) * alpha + ema[i-1]

plt.figure(figsize=(12, 6))
plt.plot(stock_data.index, closing_prices, label="Close Price")
plt.plot(stock_data.index, sma, label="SMA")
plt.plot(stock_data.index, ema, label="EMA")
plt.title(f"{stock_symbol} Moving Averages")
plt.xlabel("Date")
plt.ylabel("Price")
plt.legend()
plt.show()
```

```python
import yfinance as yf
import numpy as np
import matplotlib.pyplot as plt

def su(a,i):
    s=0
    for x in range(i+1) :
        s+=a[x]
    return s

sbi=yf.Ticker('SBIN.NS')

df = sbi.history(start="2023-01-01",end ="2023-04-01")
df2 = df[['Close']]
c=df2.to_numpy()
sma=[]
day=[]
for j in range(0,59):
    day.append(j)

for i in range (0,59):
    sma.append((su(c,i+3)-su(c,i))/3)

fig,axes=plt.subplots()

axes.plot(day,sma,'g')
axes.set_xlabel('Day')
axes.set_ylabel('Simple moving average')
plt.show()
```
