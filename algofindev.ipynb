```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import yfinance as yf
stock_symbol = "TCS.NS"
start_date = "2023-02-18"
end_date = "2023-05-18"

data = yf.download(stock_symbol, start=start_date, end=end_date)
sma = data['Close'].rolling(window=3).mean()
ema = data['Close'].ewm(span=3, adjust=False).mean()
plt.figure(figsize=(10, 6))
plt.plot(data['Close'], label='Close Price')
plt.plot(sma, label='Simple Moving Average (SMA)')
plt.plot(ema, label='Exponential Moving Average (EMA)')
plt.xlabel('Date')
plt.ylabel('Price')
plt.title('Moving Average Indicator')
plt.legend()
plt.show()

```

```python
import yfinance as yf
import numpy as np
import matplotlib.pyplot as plt

def su(a,i):
    s=0
    for x in range(i+1) :
        s+=a[x]
    return s

sbi=yf.Ticker('SBIN.NS')

df = sbi.history(start="2023-01-01",end ="2023-04-01")
df2 = df[['Close']]
c=df2.to_numpy()
sma=[]
day=[]
for j in range(0,59):
    day.append(j)

for i in range (0,59):
    sma.append((su(c,i+3)-su(c,i))/3)

fig,axes=plt.subplots()

axes.plot(day,sma,'g')
axes.set_xlabel('Day')
axes.set_ylabel('Simple moving average')
plt.show()
```
